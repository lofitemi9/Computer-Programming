<h2 style="color: teal;">PYTHON OPERATOR</h2>

---

- Exponent : \*\*
- Modulus : %
- Floor Division : // (This returns only the integer part of a division completely removing the digits after the decimal point)
- Division : /
- Multiplication: \*
- Subtraction : -
- Addition : +

---

---

<h2 style="color: teal;">ORDER OF OPERATIONS</h2>
---

1. ()
2. \*\*
3. \*, /, //, % (L TO R)
4. +, - (L TO R)

This function waits for the user to type some text on the keyboard and press ENTER

<h2 style="color: teal;">Len Function</h2>

This function returns an integer value of the number of characters in a string.


```python
firstName = 'bill'
print(len(firstName))
```

<h2 style="color: teal;">Casting Values</h2>

This means changing an integer to a string or string to an integer


```python
print("Hello " + str(myName))

testNum = 23
testNum = str(testNum)
print("Number is " + testNum)

```
