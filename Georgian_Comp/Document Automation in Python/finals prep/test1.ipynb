```python
testSentence = input("Enter a sentence: ")
sentenceLength = len(testSentence)

def letterCounter(sentence):
    letterCount = 0
    for char in sentence:
        if (char == 'e' or char == 'E'):
            letterCount += 1
    print("THE NUMBER OF occurrences of the LETTER E IS:" + str(letterCount))
            
letterCounter(testSentence)
```

```python
import re
regex = re.compile(r'\d{3}-\d{3}-\d{4}')

phoneNumber = input("Enter a phone number (format: xxx-xxx-xxxx): ")
matchObject = regex.search(phoneNumber)
if matchObject:
    print("Valid phone number format.")
else:
    print("Invalid phone number format.")
    
```
