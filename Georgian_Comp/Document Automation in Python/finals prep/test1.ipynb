```python
testSentence = input("Enter a sentence: ")
sentenceLength = len(testSentence)

def letterCounter(sentence):
    letterCount = 0
    for char in sentence:
        if (char == 'e' or char == 'E'):
            letterCount += 1
    print("THE NUMBER OF occurrences of the LETTER E IS:" + str(letterCount))
            
letterCounter(testSentence)
```

```python
import re
regex = re.compile(r'\d{3}-\d{3}-\d{4}')

phoneNumber = input("Enter a phone number (format: xxx-xxx-xxxx): ")
matchObject = regex.search(phoneNumber)
if matchObject:
    print("Valid phone number format.")
else:
    print("Invalid phone number format.")
    
```

```python
import os

firstName = ''
lastName = ''
employeeID = ''
dateOfBirth = ''

def captureInput():
    global firstName, lastName, employeeID, dateOfBirth
    firstName = input("Enter First Name: ")
    lastName = input("Enter Last Name: ")
    employeeID = input("Enter Employee ID: ")
    dateOfBirth = input("Enter Date of Birth (YYYY-MM-DD): ")

captureInput()

stringToWrite = f"First Name: {firstName}\nLast Name: {lastName}\nEmployee ID: {employeeID}\nDate of Birth: {dateOfBirth}\n"

if (os.path.exists("employeeData.txt")==False):
    testFile = open("employeeData.txt", "w")
    testFile.write(stringToWrite)
    testFile.close()
else:
    testFile = open("employeeData.txt", "w")
    testFile.write(stringToWrite)
    testFile.close()


def readFile():
    testFile = open("employeeData.txt", "r")
    fileContents = testFile.read()
    print("Employee Data:\n" + fileContents)
    testFile.close()

readFile()
```

```python
import datetime

christmas = datetime.date(2025, 12, 25)
print("Christmas is on: " + str(christmas))

today = datetime.date.today()
print("Today's date is: " + str(today)) 

today2 = datetime.datetime.now()
print("Current date and time is: " + str(today2)) 

timeDelta = christmas - today
print("Days until Christmas: " + str(timeDelta.days))

if christmas < today:
    print("Christmas has already passed this year.")
else:
    print("Christmas is yet to come this year.")
    

```
