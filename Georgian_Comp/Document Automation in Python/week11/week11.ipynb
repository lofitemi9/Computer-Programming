- Write a python application which will build a report in Word based upon the information in the spreadsheet. The first paragraph in the report should be the longest paragraph in the spreadsheet which you will determine programatically. The second paragraph in the report should be the middle paragraph(in length) in the spreadsheet which you will determine programatically. The last paragraph in the report should be the shortest paragraph in the spreadsheet which you will determine programatically. Each paragraph should have a heading, the first heading should contain the word "fish" which you will determine programatically, the second heading should contain the word "cheese" which you will determine programatically and the last heading should contain the word "car" which you will determine programatically, therefore you must ensure that the headings written into the spreadsheet should contain these words. Finally add an image of a python to the bottom of the report. If you wish to spread your skills further, underline any word in the report that ends in the letter 'r'.

```python
import openpyxl
import docx

# workbook and sheet
sampleWorkbook = openpyxl.load_workbook('C:\\xampp\\htdocs\\myProjects\\Web Development Projects\\Georgian_Comp\\Document Automation in Python\\week11\\docs\\Lab.xlsx')
sheet = sampleWorkbook.active

doc = docx.Document()

columnList = ["A", "B", "C"]


def findWordfromHeadingandAddtoDoc():
    orderList = []
    "Find heading value 'word' in the sheet and add it as a heading in the doc."
    for column in columnList:
        # iterate through all rows in the sheet 
        for row in range(1, 3):
            cellContents = sheet[column + str(row)]
            string = cellContents.value

            # ensure it's a string before calling .upper()
            if string == 'FISH':
                print("Found heading: " + string + "\nAdding to doc now.")
                # add to list
                orderList.insert(0, string)
                
            elif string == 'CHEESE':
                print("Found heading: " + string + "\nAdding to doc now.")
                # add to list
                orderList.insert(1, string)
                
            elif string == 'CAR':
                print("Found heading: " + string + "\nAdding to doc now.")
                # add to list
                orderList.insert(2, string)
                
    return orderList


# now i am finding the paragraph order according to the lengths from longest to shortest
def paragraphOrder():
    lengthsAndParagraphs = []
    
    for column in columnList:
        cellContents = sheet[column + "2"]
        string = cellContents.value
    
        paraLength = len(string)
        lengthsAndParagraphs.append((paraLength, string))


    # return ordered paragraphs
    orderedParagraphs = [para for length, para in lengthsAndParagraphs]
    return orderedParagraphs


# use a variable docx_path if present in the notebook, otherwise fallback

docx_path = r"C:\xampp\htdocs\myProjects\Web Development Projects\Georgian_Comp\Document Automation in Python\week11\docs\Lab.docx"

# add heading(s)
orderHeadings = findWordfromHeadingandAddtoDoc()

# add paragraphs sorted by length
orderedParagraphs = paragraphOrder()


if orderedParagraphs:
    # add the heading
    doc.add_heading(orderHeadings[0], level=1)
    # add the longest paragraph first
    doc.add_paragraph(orderedParagraphs[0])
    # add second heading
    doc.add_heading(orderHeadings[1], level=1)
    # add middle paragraph and third header
    if len(orderedParagraphs) >= 3:
        midIndex = len(orderedParagraphs) // 2
        doc.add_paragraph(orderedParagraphs[midIndex])
    # add third heading
    doc.add_heading(orderHeadings[2], level=1)
    # add shortest paragraph and third header
    if len(orderedParagraphs) >= 2:
        doc.add_paragraph(orderedParagraphs[-1])
        
#ad a picture
doc.add_picture("C:\\xampp\\htdocs\\myProjects\\Web Development Projects\\Georgian_Comp\\Document Automation in Python\\week11\\docs\\fish.png" , width=docx.shared.Inches(4))

# save the document
doc.save(docx_path)
```
