<h2 style="color: teal;">Comparison Operator</h2>

- Equals TO : == ; this checks whether two values are the same as each other
- Assignment : = ; this assigns the value on the right into the variable on the left

    <h4 style="color: lightyellow;">Examples</h4>

  1. . == equals to
  2. . != not equals to
  3. . < less than
  4. . > greater than
  5. . <= less than or equals to
  6. . >= greater than or equals to


```python
print(42 == 42)
print(2 != 3)
print(1 > 3)
print('hello' == 'hello')
print('hello' == "Hello")
print(42 == 42.0)
print(42 == '42')
```

<h2 style="color: teal;">Loop Logic</h2>

<h4 style="color: lightyellow;">Break Statement</h4>

This keyword makes the program exit the loop in which it was encountered in


```python
variableOne = 7
while(variableOne < 20):
    print(variableOne)
    variableOne = variableOne + 1
    if(variableOne == 10):
        break
```

<h4 style="color: lightyellow;">Continue Statement</h4>

This keyword makes python jump back to the start of the loop and reevaluate the loops condition statements


```python
variableOne = 7
while(variableOne < 15):
    variableOne = variableOne + 1
    if(variableOne == 10):
        continue
    print(variableOne)
```

<h2 style="color: teal;">For Loops / Range Function</h2>

<h4 style="color: lightyellow;">For Loops</h4>

The for loop makes the program iterate/loop over a section of code a specified number of times.


```python
for i in range(5):
    print('Jimmy Five Times (' + str(i) + ')')
```


<h4 style="color: lightyellow;">Range Function</h4>

The range function determines a range between two values (start range, end range)


```python
for i in range(12 , 16):
    print(i)
```
