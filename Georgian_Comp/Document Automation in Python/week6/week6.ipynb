```python
# 1. inputStr()
import pyinputplus as pyip
import re

def inputStr():
    word = pyip.inputStr(prompt="Enter a word: ")
    print("You typed:", word)

inputStr()
```

```python
# 2. inputStr() with allowRegexes
def inputStrAllowRegexes():
    capitalizedName = pyip.inputStr(
        prompt="Enter a Capitalized Name: ",
        allowRegexes=[r'^[A-Z][a-zA-Z]*$']
    )
    print("Accepted:", capitalizedName)

inputStrAllowRegexes()
```

```python
# 3. inputStr() with blockRegexes
def inputStrBlockRegexes():
    lettersOnly = pyip.inputStr(
        prompt="Enter a username (letters only, no spaces or digits): ",
        blockRegexes=[r'.*\d.*', r'.*\s.*']
    )
    print("Username accepted:", lettersOnly)

inputStrBlockRegexes()
```

```python
#Combination | inputStr() with both allow and block
def inputStrBlockAllowRegexes():
    word = pyip.inputStr(
        prompt= "Capitalized First Letter Name only: ",
        allowRegexes=[r'^[A-Z][a-zA-Z]*$'],
        blockRegexes=[(r'.*', 'Only Capitalized names allowed')]
    )
    print("Username accepted:", word)
    
inputStrBlockAllowRegexes()
```

```python
# 7. inputNum() with greaterThan
def inputNumGreaterThan():
    num = pyip.inputNum(prompt="Enter a number > 5: ", greaterThan=5)
    print("OK:", num)

inputNumGreaterThan()
```

```python
# 8. inputNum() with lessThan
def inputNumLessThan():
    num = pyip.inputNum(prompt="Enter a number < 50: ", lessThan=50)
    print("OK:", num)

inputNumLessThan()
```

```python
# 9. inputNum() with limit of 3
def inputNumLimit():
    try:
        num = pyip.inputNum(prompt="Enter a number between 1 and 10 (3 attempts): ",
                            min=1, max=10, limit=3)
        print("OK:", num)
    except pyip.RetryLimitException:
        print("Out of attempts. Please re-run the program.")

inputNumLimit()
```

```python
# 10. inputNum() with a timeout of 3
def inputNumTimeout():
    try:
        num = pyip.inputNum(prompt="Enter a number (you have 3 seconds): ", timeout=3)
        print("OK:", num)
    except pyip.TimeoutException:
        print("Timed out. Please re-run and answer faster.")

inputNumTimeout()
```

```python
#Part B: regex pattern finder
def regexPattern():
    def matchesPattern(text):
        pattern = re.compile(r'\d{2}[A-Z]{2}[a-z][^A-Za-z0-9]')
        return bool(pattern.search(text))

    samples = ["33ABd%", "12ZZq!", "A33ABd%", "99AAz#", "42abC!"]
    for i in samples:
        print(f"{i:12} ->", matchesPattern(i))
        
regexPattern()
```
