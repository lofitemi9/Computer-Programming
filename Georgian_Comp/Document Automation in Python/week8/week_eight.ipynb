## urlib.request

- urlopen() - open a webpage
- getcode() - pulls a status code from a server
- 200 - request succesful
- 301 - website moved permananetly
- 403 - access forbidden
- 404 - resource not found
- read() - reads the webpage


```python
import urllib.request
from bs4 import BeautifulSoup
weburl = urllib.request.urlopen('https://www.google.com')
print("Result code: " + str(weburl.getcode()))
data = weburl.read()

soup = BeautifulSoup(data, 'html.parser')
print(soup.prettify())
```

```python
import urllib.request as url
from bs4 import BeautifulSoup

# Name: Temiloluwa Shokunbi
# Student Number: 200632787
# Date: 11/7/2025
# Time: 12:55PM
# Description: This code scrapes websites for phone numbers and returns it to the user.

#starting the function and passing one parameter
def findPhoneNumbers(validURL):
    weburl = url.urlopen(validURL)
    #opens the url that is stored in validURL
    soup = BeautifulSoup(weburl.read(), 'html.parser')
    #passes url data into beautiful soup to become readable data
    text = soup.get_text()
    
    
    phoneNumbers = text.split()
    # creates a list from the outputted text data based on spacing
    for word in phoneNumbers:
        #phone number logic to scrape it from webpages
        digits = '0123456789'
        count = 0
        for i in range (len(word)):
            if word[i] in digits:
                count += 1
        if count == 10:
            print(word)
        else:
            pass
#passing a website into the function
findPhoneNumbers('https://www.barrie.ca/')
```
