## PROOF OF CONCEPT CODE


```python
import openpyxl, re
pattern = {r"\D\D\d\d\D\D"}
regexObject = re.compile(pattern)

sampleWorkbook = openpyxl.load_workbook('example.xlsx')
sheet = sampleWorkbook.active

columnList = ["B", "C", "D", "E"]
for column in columnList:
    for row in range(4, 1004):
        cellContents = sheet[str(column) +str(row)]
        string = (cellContents.value)
        matchObject = regexObject.search(string)
        if matchObject:
            print(cellContents.value + " found in cell " + str(column) + str(row))
            
    sampleWorkbook.close()
```

- Using only concepts from lecture 1-8 create a spreadsheet that contains 500 words spread across 5 columns (100 words per column), Write a functuon that takes one argument (the path of the spreadsheet). The purpose of the function is to read the spreadsheet and find any words that do not contain consonants, once found print the word and it scoordinates. Return 0 to the caller once the function has read the entire spreadsheet

- Write a caller that uses input validation to secure a proper path from the user, once acquired invoke the function and report its success to the user once it has read the entire spreadsheet


```python
import openpyxl
import re
import pyinputplus as pyip

# Name: Temiloluwa Shokunbi
# Student Number: 200632787
# Date: 11/14/2025
# Time: 12:36 PM
# Description: This script searches an Excel workbook for specific patterns in cells and reports their locations.

#this is the path that leads to the excel file

#defining our function that takes in the file path as an argument
def searchWordsWithoutConsonants(filePath):
    #passing the file path to openpyxl to load the workbook
    sampleWorkbook = openpyxl.load_workbook(filePath)


    sheet = sampleWorkbook.active
    #defining a pattern to search for consonants
    pattern = (r"[BCDFGHKLMNPQRSTVWXYZbcdfghjklmnpqrstvwxyz]")
    regexObject = re.compile(pattern)
    
    #selecting the columns and rows to search through
    columnList = ["A", "B", "C", "D", "E"]
    for column in columnList:
        #iterating through rows 1 to 100
        for row in range(1, 101):
            #accessing the cell contents
            cellContents = sheet[str(column) + str(row)]
            string = str(cellContents.value)
            matchObject = regexObject.search(string)
            #if no consonants are found in the cell, print the cell value and its coordinates
            if not matchObject:
                print(cellContents.value + " found in cell " + str(column) + str(row))
                
    sampleWorkbook.close()
    return 0

#collecting input for file path and validating it

#using pyinputplus to ensure the user inputs a valid file path

inputFilePath = pyip.inputFilepath(prompt="Please enter the file path to the Excel workbook: " )
searchWordsWithoutConsonants(inputFilePath)
```
